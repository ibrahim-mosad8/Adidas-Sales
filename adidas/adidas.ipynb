```python
df['Price per Unit'] = df['Price per Unit'].str.replace("$", "", regex= False).astype(float)
```

```python
df['Price per Unit'].dtype
```

```python
df['Units Sold'] = pd.to_numeric(df['Units Sold'], errors = "coerce").astype("Int64")
```

```python
df['Units Sold'].dtype
```

```python
df["Total Sales"] = pd.to_numeric(
    df["Total Sales"].str.replace(",", "", regex=False)
).astype(int)
```

```python
df['Total Sales'].dtype
```

```python
df['Operating Profit'] = pd.to_numeric(
    df['Operating Profit']
        .astype(str)
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.strip(),
    errors="coerce"
)
```

```python
df['Operating Profit'].dtype
```

```python
df.head()
```

```python
df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])
```

```python
df.info()
```

# Total Sales by Region 

```python
Total_sales_by_Region = pd.pivot_table(df,
                                        values = "Total Sales"
                                        ,index= "Region"
                                        ,aggfunc = "sum")
Total_sales_by_Region
```

# Average Operating Profit per Product

```python
AVG_operation_profit= pd.pivot_table(df,
                                      values="Operating Profit",
                                      index = "Product",
                                       aggfunc = "mean")
AVG_operation_profit
```

# Unit Solid by Sales Method

```python
Unit_Sold_by_method = pd.pivot_table(df,
                                    values = "Units Sold",
                                    index = "Sales Method",
                                    aggfunc = "sum")
Unit_Sold_by_method
```

# Monthly Total Sales by Region

```python
df["month"] = df["Invoice Date"].dt.to_period("M")
monthly_totalSales = pd.pivot_table(df,
                  values = "Total Sales",
                  index = "month",
                  columns = "Region",
                  aggfunc = "sum")
monthly_totalSales
```

# Top 10 Retailers by Total Sales

```python
Top10Retailers = pd.pivot_table(df,
                               values = "Total Sales",
                               index = "Retailer",
                               aggfunc = "sum").sort_values(by="Total Sales", ascending = False)
Top10Retailers
```

# Total Sales Per Product

```python
TotalSales = df.groupby("Product")["Price per Unit"].mean()
TotalSales
```

# Total Units Sold per City

```python
TotalUnitSold = df.groupby("City")["Units Sold"].sum().sort_values(ascending = False)
TotalUnitSold.head(10)
```

# Average Price per Unit by Region

```python
AVG_price_per_unit = df.groupby("Region")["Price per Unit"].mean()
AVG_price_per_unit
```

# Total Operating Profit per Sales Method

```python
Total_Operating_Profit = df.groupby("Sales Method")["Operating Profit"].sum()
Total_Operating_Profit
```

# Number of Invoices per Retailer

```python
NumInvoicesRetailer = df.groupby("Retailer")["Invoice Date"].count().sort_values(ascending = False)
NumInvoicesRetailer
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x = Total_sales_by_Region.index,
           y = Total_sales_by_Region["Total Sales"])
plt.title("Top Regions by Total Sales - like I tracked them myself")
plt.ylabel("Total Sales")
plt.xlabel("Region")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=AVG_operation_profit.index,
           y=AVG_operation_profit["Operating Profit"])
plt.xticks(rotation=45)
plt.title("Average Operating Profit per Product ")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df.groupby("Sales Method")["Units Sold"].sum().plot.pie(autopct = "%1.1f%%")
plt.title("Units Sold by Sales Method")
plt.ylabel("")
plt.show()
```

```python
monthly_totalSales.plot(figsize=(12,6), marker= "o")
plt.title("Monthly Sales by Region")
plt.ylabel("Month")
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.histplot(df["Price per Unit"], bins=20,kde=True)
plt.title("Distribution of Price per Unit")
plt.show()
```

```python
plt.figure(figsize =(8,5))
sns.scatterplot(x="Units Sold", y= "Total Sales", data = df, hue="Region")
plt.title("Units Sold vs Total Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x= "Region", y= "Operating Profit", data= df)
plt.title("Operating Profits Spread by Region")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.barplot(x="Product", y="Total Sales", data= df, estimator=sum)
plt.xticks(rotation = 45)
plt.title("Total Sales by Product")
plt.show()
```

```python
top_cities = df.groupby("City")["Units Sold"].sum().sort_values(ascending = False).head(10)
plt.figure(figsize = (8,5))
sns.barplot(x=top_cities.index,
           y=top_cities.values)
plt.xticks(rotation=45)
plt.title("Top 10 Cities by Units Sold")
plt.show()
```
